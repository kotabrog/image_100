```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
import os
import sys
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)

from image_100 import *
```

```python
img_orig = io.imread("./cat-323262_1920.jpg")

plt.imshow(img_orig)
plt.show()
```

## rgb_to_hsv

RGBをHSVに変換する (h: 0-360, s: 0-128, v:0-256)

```python
img_hsv = hsv.rgb_to_hsv(img_orig)
```

## hsv_to_rgb

HSVをRGBに変換する (h: 0-360, s: 0-128, v:0-256)

```python
img_from_hsv = hsv.hsv_to_rgb(img_hsv)
print("orig == from_hsv ? ", np.all(img_orig == img_from_hsv))
show_img.show_img_origin_and_conversion(img_orig, img_from_hsv)
```

## turn_hsv

色相(Hue)を反転させる

```python
img_from_hsv2 = hsv.turn_hsv(img_orig)
show_img.show_img_origin_and_conversion(img_orig, img_from_hsv2)
```
